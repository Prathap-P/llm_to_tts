```python
import sounddevice as sd

# Load model
tts = TTS(
    model_name="tts_models/multilingual/multi-dataset/xtts_v2",
    gpu=False
)

# Generate speech (NumPy array)
wav = tts.tts(
    text="Hello world. This audio is generated and played immediately.",
    language="en"
)

# Play audio
sd.play(wav, samplerate=tts.synthesizer.output_sample_rate)
sd.wait()
```
