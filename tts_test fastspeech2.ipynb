### Fastspeech2 example

```python
from espnet2.bin.tts_inference import Text2Speech
import numpy as np
import soundfile as sf
```

```python
fastspeech_model = Text2Speech.from_pretrained(
    "espnet/kan-bayashi_ljspeech_fastspeech2"
)
```

```python
def split_into_sentences(text):
    """Split text into sentences for better prosody"""
    import re
    sentences = re.split(r'(?<=[.!?])\s+', text)
    return [s.strip() for s in sentences if s.strip()]

def synthesize_long_text(text, output_file="output.wav"):
    """Synthesize long text by processing sentence by sentence"""
    sentences = split_into_sentences(text)
    
    audio_chunks = []
    
    for i, sentence in enumerate(sentences):
        print(f"Processing sentence {i+1}/{len(sentences)}")
        
        # Generate audio for this sentence
        speech = fastspeech_model(sentence, speed_control_alpha=1.2)["wav"]
        audio_chunks.append(speech.numpy())
        
        # Add small pause between sentences (adjust as needed)
        pause = np.zeros(int(fastspeech_model.fs * 0.3))  # 0.3 second pause
        audio_chunks.append(pause)
    
    # Concatenate all audio
    full_audio = np.concatenate(audio_chunks)
    
    # Save to file
    sf.write(output_file, full_audio, fastspeech_model.fs)
    print(f"Saved to {output_file}")
    
    return full_audio
```
